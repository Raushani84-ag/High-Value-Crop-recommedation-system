```python
def expand_with_crops(df, crop_mapping):
    rows = []

    for _, row in df.iterrows():
        district = row['district']
        season = row['Season']

        # Get crops for that district + season
        crops = crop_mapping.get(district, {}).get(season, [])

        for crop in crops:
            new_row = row.copy()
            new_row['target_crop'] = crop
            rows.append(new_row)

    return pd.DataFrame(rows)
```
